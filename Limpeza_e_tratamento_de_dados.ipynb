```python
import pandas as pd
import json
dados_churn = pd.read_json("dataset-telecon.json")
```

```python
dados_churn.head(5)
```

**Transformando o objeto JSON**

```python
with open ("/content/dataset-telecon.json") as f:
  json_bruto = json.load(f)
```

**Normalizando o objeto Json**

```python
dados_normalizados = pd.json_normalize(json_bruto)
dados_normalizados.head(5)
```

```python
dados_normalizados.info()
```

***Como podemos verificar acima, algumas colunas constam com dados nulos. Iremos realizar o ajuste necessário para corrigir este problema. Outro detalhe que devemos alterar é a coluna conta.cobranca.Total que está com o type errado de dados, existindo apenas um espaço em vazio em toda a coluna.***

```python
dados_normalizados[dados_normalizados["conta.cobranca.Total"] == " "][
    ['cliente.tempo_servico', 'conta.contrato', 'conta.cobranca.mensal', 'conta.cobranca.Total']
]
```

***Aqui podemos verificar que a coluna conta.contrato possui apenas um valor "dois anos" que é igual a 24 meses. Dessa forma, se realizarmos a multiplicação da coluna conta.contrato pela coluna conta.cobranca.mensal iremos obter o valor da coluna conta.cobranca.Total. Faremos esta operação informando o índice de cada linha, para evitar uma operação no conjunto de dados inteiro. Para pegar o índice, copiaremos o comando***

```python
idx = dados_normalizados[dados_normalizados['conta.cobranca.Total'] == ' '].index
```

***Percorremos cada indice da coluna conta.cobranca.Total e realizamos a operação de multiplicação para cada valor na coluna conta.cobranca.mensal. Por último associamos o resultado a coluna conta.cobranca.Total e substituímos o valor dos indices da coluna cliente.tempo_serviço.***


```python
dados_normalizados.loc[idx, "conta.cobranca.Total"] = dados_normalizados.loc[idx, "conta.cobranca.mensal"] * 24
```

```python
dados_normalizados.loc[idx, "cliente.tempo_servico"] = 24
```

```python
dados_normalizados.info()
```

```python
for col in dados_normalizados.columns:
  print(f"Coluna: {col}")
  print(f"Valores: {dados_normalizados[col].unique()}")
  print('-'*30)
```
